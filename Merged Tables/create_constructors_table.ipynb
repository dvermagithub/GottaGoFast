```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as scipy
import statsmodels.api as sm
%matplotlib inline
```

```python
#Import Data
circuits = pd.read_csv('circuits.csv')
constructor_results = pd.read_csv('constructor_results.csv')
constructor_standings = pd.read_csv('constructor_standings.csv')
constructors = pd.read_csv('constructors.csv')
driver = pd.read_csv('driver.csv')
driver_standings = pd.read_csv('driver_standings.csv')
lap_times = pd.read_csv('lap_times.csv')
pit_stops = pd.read_csv('pit_stops.csv')
qualifying = pd.read_csv('qualifying.csv')
races = pd.read_csv('races.csv')
results = pd.read_csv('results.csv')
status = pd.read_csv('status.csv')
```

```python
# Merging Constructor_standings with Constructors (ignoring constuructor_results as it conatins same columns)
constructor_int1 = constructor_standings.merge(constructors, on='constructorId', how='inner')
```

```python
display(constructor_int1)
```

```python
#Merging above with Races to add race details
constructor_int2 = constructor_int1.merge(races, on='raceId', how='inner')
```

```python
display(constructor_int2)
```

```python
#Merging above with Circuits to add circuit details
constructor_int3 = constructor_int2.merge(circuits, on='circuitId', how='inner')
```

```python
display(constructor_int3)
```

```python
#Dropping the columns not needed:
constructors_full = constructor_int3.drop(columns=['constructorStandingsId' , 'positionText', 'name_x', 'url_x', 'time', 'url_y', 'name', 'alt', 'url' ])
```

```python
constructors_full.to_csv('constructors_full.csv')
```
