```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as scipy
import statsmodels.api as sm
%matplotlib inline
```

```python
#Import Data
circuits = pd.read_csv('circuits.csv')
constructor_results = pd.read_csv('constructor_results.csv')
constructor_standings = pd.read_csv('constructor_standings.csv')
constructors = pd.read_csv('constructors.csv')
driver = pd.read_csv('driver.csv')
driver_standings = pd.read_csv('driver_standings.csv')
lap_times = pd.read_csv('lap_times.csv')
pit_stops = pd.read_csv('pit_stops.csv')
qualifying = pd.read_csv('qualifying.csv')
races = pd.read_csv('races.csv')
results = pd.read_csv('results.csv')
status = pd.read_csv('status.csv')
```

```python
# Merging Results with Status
results_int1 = results.merge(status, on='statusId', how='inner')
results_int1
```

```python
# Merging Results with Driver Information
results_int2 = results_int1.merge(driver, on='driverId', how='inner')
results_int2
```

```python
# Merging Results with Race Information
results_int3 = results_int2.merge(races, on='raceId', how='inner')
results_int3
```

```python
# Merging Results with Circuit Information
results_int4 = results_int3.merge(circuits, on='circuitId', how='inner')
results_int4
```

```python
# Merging Results with Constructor Information
results_int5 = results_int4.merge(constructors, on='constructorId', how='inner')
results_int5
```

```python
# Merging Results with Qualification Information
results_int6 = results_int5.merge(qualifying, on=['raceId', 'driverId'], how='outer', suffixes=('','_quali'))
results_int6
```

```python
#Dropping the columns not needed:
results_full = results_int6.drop(columns=['number_x' , 'position', 'time_x', 'milliseconds', 'fastestLap', 'rank',
                                          'statusId', 'number_y', 'code', 'forename','surname','url_x', 'circuitId',
                                          'name_x', 'time_y', 'url_y', 'name_y', 'lat', 'lng', 'alt', 'url_x', 'name',
                                          'qualifyId', 'constructorId_quali', 'number', 'q1', 'q2', 'q3'])
```

```python
results_full.to_csv('results_full.csv')
```
